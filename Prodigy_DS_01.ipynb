```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
from matplotlib import style
```

```python

df = pd.read_csv("C:\\Users\\user\\Desktop\\Prodigy InfoTech\\population.csv")

df
```

```python
saarc_countries = ['Afghanistan', 'Bangladesh', 'Bhutan', 'India', 'Maldives', 'Nepal', 'Pakistan', 'Sri Lanka']


df_saarc = df[df['Country Name'].isin(saarc_countries)][['Country Name', '2023']]

style.use('ggplot')
plt.figure(figsize=(10,5))
color = ['royalblue', 'red', 'purple', 'yellow', 'lime', 'orange', 'aqua', 'violet']

df_saarc.plot(kind='bar', x='Country Name', y='2023', color=color, legend=False)
plt.yscale('log')
plt.title('Population of the SAARC countries in 2023')
plt.xlabel('Country')
plt.ylabel('Population')
plt.show()
```
