Histogram of Population of the World

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
from matplotlib import style
```

```python

df = pd.read_csv("C:\\Users\\user\\Desktop\\Prodigy InfoTech\\population.csv")

df
```

```python
saarc_countries = ['Afghanistan', 'Bangladesh', 'Bhutan', 'India', 'Maldives', 'Nepal', 'Pakistan', 'Sri Lanka']


df_saarc = df[df['Country Name'].isin(saarc_countries)][['Country Name', '2023']]

style.use('ggplot')
plt.figure(figsize=(10,5))
color = ['royalblue', 'red', 'purple', 'yellow', 'lime', 'orange', 'aqua', 'violet']

df_saarc.plot(kind='bar', x='Country Name', y='2023', color=color, legend=False)
plt.yscale('log')
plt.title('Population of the SAARC countries in 2023')
plt.xlabel('Country')
plt.ylabel('Population')
plt.show()
```

```python
# Select only the columns related to years (i.e., 1960 to 2023)
population_data = df.loc[:, '1960':'2023'].values.flatten()  # Flatten to convert to a 1D array

# Ensure the data is numeric and remove any NaN values
population_data = pd.to_numeric(population_data, errors='coerce')
population_data = population_data[~pd.isna(population_data)]

# Plot the histogram
plt.figure(figsize=(10, 5))
plt.hist(population_data, bins=10, color= "aqua", edgecolor='black')

# Add title and labels
plt.title('Histogram of Population Data of India between (1960-2023)')
plt.yscale('log')
plt.xlabel('Population')
plt.ylabel('Frequency')
# Show the plot
plt.show()
```
