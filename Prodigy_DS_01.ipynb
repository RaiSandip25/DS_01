Bar Graph and Histogram of Population of the World

```python
# Importing different Python Libraries

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Importing style from matplotlib

from matplotlib import style
```

```python
#Load the population data from a CSV file into a pandas DataFrame

df = pd.read_csv("C:\\Users\\user\\Desktop\\Prodigy InfoTech\\population.csv")

#Display the DataFrame
df
```

```python
# Define a list of SAARC countries for filtering the dataset

saarc_countries = ['Afghanistan', 'Bangladesh', 'Bhutan', 'India', 'Maldives', 'Nepal', 'Pakistan', 'Sri Lanka']

# Filter the DataFrame to include only rows where 'Country Name' is in the SAARC countries list
# Select only the 'Country Name' and '2023' columns for visualization

df_saarc = df[df['Country Name'].isin(saarc_countries)][['Country Name', '2023']]

style.use('ggplot')   #styling the barplot

plt.figure(figsize=(20,10))   #setting the figure size
color = ['royalblue', 'red', 'purple', 'yellow', 'lime', 'orange', 'aqua', 'violet']   #defining the color palette

df_saarc.plot(kind='bar', x='Country Name', y='2023', color=color, legend=False)  #Creating bar plot

plt.title('Population of the SAARC countries in 2023')    # Adding Title to the plot
plt.xlabel('Country')      # Label the x-axis
plt.ylabel('Population')   # Label the y-axis

# Set the y-axis to a logarithmic scale to better visualize differences in population sizes

plt.yscale('log')

# Display the plot

plt.show()
```

```python
# Filter the DataFrame for India
df_india = df[df['Country Name'] == 'India']

df_india
```

```python
# Select only the columns related to years (i.e., 1960 to 2023)
population_data = df_india.loc[:, '1960':'2023'].values.flatten()  # Flatten to convert to a 1D array

# Ensure the data is numeric and remove any NaN values
population_data = pd.to_numeric(population_data, errors='coerce')
population_data = population_data[~pd.isna(population_data)]

# Plot the histogram
plt.figure(figsize=(10, 5))
plt.hist(population_data, bins=20, color= "aqua", edgecolor='black')

# Add title and labels
plt.title('Histogram of Population Data of India between (1960-2023)')
plt.yscale('log')
plt.xlabel('Population')
plt.ylabel('Frequency')
# Show the plot
plt.show()
```
